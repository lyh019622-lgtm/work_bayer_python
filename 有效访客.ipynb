```python
# 1. 定义文件路径（根据你的实际路径修改！）
file_path = r"D:\rpa\有效访客数\收集.txt"  # Windows路径（r表示原生字符串，避免转义）
# file_path = "/Users/xxx/data/nums.txt"  # macOS/Linux路径（取消注释并修改）

# 2. 读取文件并处理
try:
    # 打开文件，读取所有内容（encoding='utf-8'适配中文/特殊字符，newline=''保留换行格式）
    with open(file_path, "r", encoding="utf-8", newline="") as f:
        file_content = f.read()  # 读取文件全部内容
    
    # 3. 处理内容：按换行分割 → 过滤空行 → 用逗号拼接
    # strip()去除首尾空白，split("\n")按换行分割成列表
    num_list = [line.strip() for line in file_content.strip().split("\n") if line.strip()]
    result = ",".join(num_list)  # 用逗号拼接成一行
    
    # 4. 输出结果（也可以写入新文件）
    print("拼接后的结果：")
    print(result)


except FileNotFoundError:
    print(f"错误：未找到文件，请检查路径是否正确 → {file_path}")
except Exception as e:
    print(f"处理出错：{e}")
```
